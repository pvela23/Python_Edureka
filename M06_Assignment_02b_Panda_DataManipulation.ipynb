```python
# 5.1: Draw a Simple plot

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = [1,2,3,4] 
y = [20, 21, 20.5, 20.8]

#plt.style.use('_mpl-gallery')
#fig, ax = plt.subplots()
#ax.plot(x, y, linewidth=2.0)

plt.plot(x,y)
plt.show()
```

```python
# 5.2: Configure the line and markers in a simple plot
# plt.subplot(facecolor='orange')
plt.plot(x,y, marker='D', color='navy')
plt.show()

# other styles
#plt.plot(x,y, color='gold')
#plt.plot(x,y,linestyle='--', marker='o', color='r')
#plt.show()
```

```python
# 5.3: configure the axes
x = [1,2,3,4]
y = [20,21,20.5,20.8]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,linestyle=':', marker='D', color='r')
plt.show()
```

```python
# 5.4: Giving title of Graph & labels of x axis and y axis

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,linestyle='--', marker='D', color='blue')

ax.set_title("Ex5.4 - Simple x and y plot")
ax.set_xlabel('Plot X')
ax.set_ylabel('Plot Y')

plt.show()
```

```python
# 5.5: Give error bar if y_error = [0.12, 0.13, 0.2, 0.1]

y_error = [0.12, 0.13, 0.2, 0.1]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,linestyle='-', marker='D', color='navy')

plt.errorbar(x, y, fmt='ro', label="data",
             yerr=y_error, ecolor='black')

ax.set_title("Ex5.4 - Simple x and y plot")
ax.set_xlabel('Plot X')
ax.set_ylabel('Plot Y')
plt.show()
```

```python
# 5.6: define width, and height as figsize=(4,5) DPI and adjust plot dpi=100

#### matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, 
#### frameon=True, FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)

y_error = [0.12, 0.13, 0.2, 0.1]

fig = plt.figure(num=None, figsize= [4, 5], dpi=100, 
                 facecolor='dodgerblue', edgecolor='red', frameon=True, clear=False)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,linestyle='-', marker='D', color='navy')

plt.errorbar(x, y, fmt='ro', label="data",
             yerr=y_error, ecolor='black')

ax.set_title("Ex5.4 - Simple x and y plot")
ax.set_xlabel('Plot X')
ax.set_ylabel('Plot Y')
plt.show()
```

```python
### 5.7: Give a font size of 14

#### matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, 
#### frameon=True, FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)

y_error = [0.12, 0.13, 0.2, 0.1]

fig = plt.figure(num=None, figsize= [4, 5], dpi=100, 
                 facecolor='dodgerblue', edgecolor='red', frameon=True, clear=False)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,linestyle='-', marker='D', color='navy')

plt.errorbar(x, y, fmt='ro', label="data",
             yerr=y_error, ecolor='black')

# Change the font size and lable to show veritical
#plt.xticks(fontsize = 14)
#plt.yticks(fontsize = 14)

ax.set_title("Ex5.4 - Simple x and y plot", fontsize = 14)
ax.set_xlabel('Plot X', fontsize = 14)
ax.set_ylabel('Plot Y', fontsize = 14)
plt.show()
```

5.9: Create a dataframe from the following data

'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],

'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],

'female': [0, 1, 1, 0, 1],

'age': [42, 52, 36, 24, 73],

'preTestScore': [4, 24, 31, 2, 3],

'postTestScore': [25, 94, 57, 62, 70]

Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age.

```python
'''
5.9: Create a dataframe from the following data
'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]
Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age.
'''

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)
print(df)

sns.scatterplot(data=df,x="preTestScore",y="postTestScore", s=100, color=".2",  palette="deep",
                sizes=(200, 2000),legend="full");
# finishing off with a semi colon (;) removed the <AxesSubplot:xlabel='preTestScore', ylabel='postTestScore'>
```

```python
## 5.10: Draw a Scatterplot from the data in question 9 of preTestScore and postTestScore with the size = 300
## and the color determined by sex.

sns.scatterplot(data=df,x="preTestScore",y="postTestScore", hue="female",s=300, 
                sizes=(200, 2000),legend="full");
```
